## 3. Feature-Level Data Quality Plan (Draft Table)

说明：该表满足 DQP 对‘每个特征都要有问题、方案、最终决策、理由、最终类型’的要求。
- `Issue Summary`: 来自 DQR
- `Candidate Solutions`: 候选方案（可多个）
- `Final Decision`: 最终采用方案（初稿允许空）

```python
# Build a DQP table template from current DQR seed
plan = dqp_seed.rename(columns={
    'Feature Type': 'Current Type',
    'Data Quality Issue': 'Issue Summary',
    'Handling Strategy': 'Candidate Solutions'
}).copy()

plan['Final Decision'] = '[TO_BE_FILLED]'
plan['Decision Rationale'] = '[TO_BE_FILLED]'
plan['Final Feature Type'] = '[TO_BE_FILLED]'
plan['Implementation Note'] = '[TO_BE_FILLED]'
plan['Validation Check'] = '[TO_BE_FILLED]'

# Pre-fill a few likely final-type hints (can still be edited manually)
type_hint = {
    'Price (€)': 'continuous',
    'Not Full Market Price': 'categorical',
    'VAT Exclusive': 'categorical',
    'Description of Property': 'categorical',
    'County': 'categorical',
    'Sale Year': 'categorical',
    'Sale Month': 'categorical',
    'Date of Sale (dd/mm/yyyy)': 'datetime',
}
for i, f in enumerate(plan['Feature']):
    if f in type_hint:
        plan.loc[i, 'Final Feature Type'] = type_hint[f]

plan
```
